```python
import numpy as np
import sympy as sy
import matplotlib.pyplot as plt

x=sy.Symbol('x')
f=sy.Function('f')
g=sy.Function('g')

vx=np.linspace(-5,5,100)

print('入力例：x**2, sin(x), log(x), exp(x)')


fx=input('fx= ')


f=sy.diff(fx)
print('微分は',f)

fx=sy.lambdify(x,fx,'numpy')
func=sy.lambdify(x,f,'numpy')
plt.plot(vx,fx(vx),label='fx')
plt.plot(vx,func(vx),label='fx_diff')
plt.grid()
plt.legend()
plt.show()
```
